```python
qvi=df_pre1.copy()
def monthly_store_metrics():
    store_yrmo_group = qvi.groupby(["STORE_NBR", "DATE"])

    
    total = store_yrmo_group["TOT_SALES"].sum()
    print(total)

    num_cust = store_yrmo_group["LYLTY_CARD_NBR"].nunique()
    
    #trans_per_cust = store_yrmo_group["TXN_ID"].nunique() / store_yrmo_group["LYLTY_CARD_NBR"].nunique()
    avg_chips_per_cust = store_yrmo_group["PROD_QTY"].sum() / num_cust
    #avg_chips_price = store_yrmo_group["TOT_SALES"].sum() / store_yrmo_group["PROD_QTY"].sum()

    
    aggregates = [total, num_cust]#,avg_chips_per_cust]#trans_per_cust,avg_chips_price
 

    metrics = pd.concat(aggregates, axis=1)
    metrics.columns = ["TOT_SALES", "nCustomers"]# ,"nChipsPerCus"]#"nTxnPerCust","avgPricePerUnit"]
    return metrics
```
